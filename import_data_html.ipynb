```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,5,8,9,10]
z=[8,7,6,3,5]
```

```python
plt.plot(x,y,color="red", lw=3, ls='--', marker='o',markersize=10,markerfacecolor='g')
plt.plot(x,z,color="red", lw=3, ls='--', marker='o',markersize=10,markerfacecolor='k')
plt.legend(['x vs y', 'x vs z'])
plt.title("my graph")
plt.xlabel("speed")
plt.ylabel("time")
plt.savefig('abc.jpg', dpi=200)
```
